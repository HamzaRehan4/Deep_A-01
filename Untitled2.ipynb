```python
import cv2
import numpy as np
from keras.applications.resnet50 import ResNet50, preprocess_input, decode_predictions

# Load the image in grayscale
directory = 'data/'
IMAGE = "subject01.centerlight"
filename = directory + IMAGE

# Load the image with error handling
try:
    pixels = cv2.imread(filename, cv2.IMREAD_GRAYSCALE)

    if pixels is None:
        raise ValueError("Image could not be loaded.")

    # Convert the grayscale image to 3 channels (RGB)
    pixels_rgb = cv2.cvtColor(pixels, cv2.COLOR_GRAY2RGB)

    # Resize the image to match the expected input size of ResNet50 (e.g., 224x224 pixels)
    desired_size = (224, 224)
    pixels_resized = cv2.resize(pixels_rgb, desired_size)

    # Add a batch dimension
    pixels_for_model = np.expand_dims(pixels_resized, axis=0)

    # Preprocess the image (apply the preprocessing used during training)
    pixels_preprocessed = preprocess_input(pixels_for_model)

    # Load the ResNet50 model
    model = ResNet50(weights='imagenet')

    # Make predictions using the preprocessed image
    predictions = model.predict(pixels_preprocessed)

    # Decode and print the top predictions
    decoded_predictions = decode_predictions(predictions, top=5)[0]
    for i, (imagenet_id, label, score) in enumerate(decoded_predictions):
        print(f"Top {i + 1}: {label} ({score:.2f})")

except Exception as e:
    print(f"Error: {str(e)}")
```
