```python
import os
import face_recognition
import numpy as np

# Directory where the employee images are stored
employee_images_dir = "data"

# Initialize variables to store face encodings and employee names
known_face_encodings = []
known_employee_names = []

# Load existing employee encodings and names
def load_employee_encodings():
    global known_face_encodings, known_employee_names
    for filename in os.listdir(employee_images_dir):
        if filename.endswith(".jpg"):
            employee_name = os.path.splitext(filename)[0]
            image_path = os.path.join(employee_images_dir, filename)
            image = face_recognition.load_image_file(image_path)
            encoding = face_recognition.face_encodings(image)[0]
            known_face_encodings.append(encoding)
            known_employee_names.append(employee_name)

# Function to update the model with a new employee
def update_model(new_employee_image_path, new_employee_name):
    global known_face_encodings, known_employee_names
    image = face_recognition.load_image_file(new_employee_image_path)
    encoding = face_recognition.face_encodings(image)[0]
    known_face_encodings.append(encoding)
    known_employee_names.append(new_employee_name)

# Function to recognize a face
def recognize_face(input_image_path):
    image = face_recognition.load_image_file(input_image_path)
    face_locations = face_recognition.face_locations(image)
    face_encodings = face_recognition.face_encodings(image, face_locations)

    for face_encoding in face_encodings:
        matches = face_recognition.compare_faces(known_face_encodings, face_encoding)
        name = "Unknown"

        if True in matches:
            first_match_index = matches.index(True)
            name = known_employee_names[first_match_index]

        print(f"Found employee: {name}")

if __name__ == "__main__":
    load_employee_encodings()

    # Example of updating the model with a new employee
    new_employee_image_path = "new_employee.jpg"
    new_employee_name = "New Employee"
    update_model(new_employee_image_path, new_employee_name)

    # Example of recognizing a face
    input_image_path = "test_image.jpg"
    recognize_face(input_image_path)
```
